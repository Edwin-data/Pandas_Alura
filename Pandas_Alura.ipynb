## **Demandas de Machine Learning**

###¿Cuáles son los valores promedio de alquiler por tipo de propiedad?



```python
promedio_alquiler_por_tipo = datos.groupby("Tipo")["Valor"].mean()
print(promedio_alquiler_por_tipo)
```

```python
promedio_alquiler_por_tipo = datos.groupby("Tipo")["Valor"].mean().sort_values(ascending=True)
print(promedio_alquiler_por_tipo)
```

```python
promedio_alquiler_por_tipo.plot(kind='barh', figsize=(12, 8), color='purple')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
promedio_alquiler_por_tipo.plot(kind='barh')
plt.title('Promedio de Alquiler por Tipo de Propiedad')
plt.xlabel('Promedio de Alquiler')
plt.ylabel('Tipo de Propiedad')
plt.tight_layout()
plt.show()
```
